```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
data
# data.head()
```

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,"Region"] = 0
data.loc[122:,"Region"] = 1
```

```python
data.info()
```

```python
data[['Region']] = data[['Region']].astype(int)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
##Dopping the null values
df = data.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##Dropping the unwanted row with index value and resetting the index values
df=df.drop(122).reset_index(drop=True)
```

```python
##Printing the new value for row with index 122
df.iloc[[122]]
```

```python
df.columns
```

```python
##fix spaces in columns name('Classes '-->'Classes')
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
##Change the required columns as integer data type
df[['day','month','year','Temperature','RH']] = df[['day','month','year','Temperature','RH']].astype(int)
```

```python
df.info()
```

```python
###Change the other columns to float type
object=[features for features in df.columns if df[features].dtypes=='O']
```

```python
##Changing the data types of columns with object('O') type to float type except classes 
for i in object:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
##Lets save the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_cleaned_dataset.csv',index=False)
```

##  Exploratory Data Analysis

```python
## Dropping day, month , year 
df_copy = df.drop(['day','month','year'],axis=1)
df_copy.head()
```

```python
## encode of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## plot desnity-plot for all features
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for a pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
## Plotting pie chart
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```
