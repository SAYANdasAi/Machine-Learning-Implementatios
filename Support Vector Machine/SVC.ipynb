### Support Vector Classifier Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Let's  create synthetic data points
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(f'{classification_report(y_test,y_pred)}')
print(cm)
sns.heatmap(cm,annot=True)
```

```python
rbf = SVC(kernel='rbf')
rbf.fit(x_train,y_train)
y_pred1 = rbf.predict(x_test)
cm = confusion_matrix(y_test,y_pred1)
print(f'{classification_report(y_test,y_pred1)}')
print(cm)
sns.heatmap(cm,annot=True)
```

```python
poly = SVC(kernel='poly')
poly.fit(x_train,y_train)
y_pred2 = poly.predict(x_test)
cm = confusion_matrix(y_test,y_pred2)
print(f'{classification_report(y_test,y_pred2)}')
print(cm)
sns.heatmap(cm,annot=True)
```

```python
sigmoid = SVC(kernel='sigmoid')
sigmoid.fit(x_train,y_train)
y_pred3 = sigmoid.predict(x_test)
cm = confusion_matrix(y_test,y_pred3)
print(f'{classification_report(y_test,y_pred3)}')
print(cm)
sns.heatmap(cm,annot=True)
```

### Hyperparameter Tuning with SVC

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']

}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4 = grid.predict(x_test)
cm = confusion_matrix(y_test,y_pred4)
print(f'{classification_report(y_test,y_pred4)}')
print(cm)
sns.heatmap(cm,annot=True)
```
